```python
# Import libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
customers = pd.read_csv('Customers.csv')
transactions = pd.read_csv('Transactions.csv')
```

```python
# Merge datasets
merged_data = transactions.merge(customers, on='CustomerID')
```

```python
# Aggregate data for clustering
customer_data = merged_data.groupby('CustomerID').agg({
    'TotalValue': 'sum',
    'Quantity': 'sum'
}).reset_index()
```

```python
# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_data[['TotalValue', 'Quantity']])
```

```python
# Apply KMeans clustering
kmeans = KMeans(n_clusters=5, random_state=42)  # Choose 5 clusters as an example
clusters = kmeans.fit_predict(scaled_data)

# Add clusters to the dataset
customer_data['Cluster'] = clusters
```

```python
# Evaluate clustering using Davies-Bouldin Index
db_index = davies_bouldin_score(scaled_data, clusters)
print(f"Davies-Bouldin Index: {db_index}")
```

```python
# Visualize clusters
sns.scatterplot(
    x=scaled_data[:, 0], y=scaled_data[:, 1], 
    hue=customer_data['Cluster'], palette='Set1'
)
plt.title('Customer Clusters')
plt.xlabel('TotalValue (Scaled)')
plt.ylabel('Quantity (Scaled)')
plt.legend(title='Cluster')
plt.show()
```
